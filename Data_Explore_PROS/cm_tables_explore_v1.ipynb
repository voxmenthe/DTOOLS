```python
headers = [x for x in os.listdir(COL_PATH) if '.head' in x]
print(headers)
```

```python
def read_table(table_name,col_path=COL_PATH,data_path=DATA_PATH):
    cols = pd.read_csv(col_path + table_name + '.head',delimiter='\n',header=None)
    cols = list(cols[0].str.replace(' ',''))
    df = pd.read_csv(data_path + table_name + '.csv',delimiter='\t',header=None,names=cols)
    return df
```

```python
tables = ['cm_cart_item_addition', 'cm_cart_item_purchase', 'cm_order', 'cm_page_view', 
          'cm_product_view', 'cm_session_first_page_view', 'cm_technical_properties']
```

```python
dfs = OrderedDict()
for table in tables:
    df = read_table(table)
    dfs[table] = df
    print("First 5 cols of table: {}".format(table),'\n')
    print(df.iloc[:3,:3].head(3))
    print()
```
